```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
import random
import os
from PIL import Image
from scipy.ndimage import zoom
import skimage
```

```python
# se cargan los conjuntos de datos de santa maria
# replace data_dir with path to the tensorflow_ds
liver_ds, liver_info = tfds.load('liver_dataset', data_dir='/media/roberto/TOSHIBA EXT/tensorflow_ds', with_info=True)
```

```python
import matplotlib.pyplot as plt

keys = list(liver_info.splits.keys())
n_patients = 10
n_samples = 1

# Inicializar contadores
patients_counter = 0

# Iterar sobre las claves comunes a los tres conjuntos de datos
for key in keys:
    patients_counter += 1
    # Obtener datos para cada modalidad del paciente actual
    patient_data = liver_ds[key]
    print("patient:", key)

    # Inicializar el contador de muestras
    samples_counter = 0

    # Iterar sobre las muestras de cada modalidad
    for sample in patient_data:
        samples_counter += 1

        # Crear una figura con subgr√°ficos
        f, axarr = plt.subplots(1, 3, figsize=(10, 7))

        img = sample['image']
        mask = sample['mask']
        
        axarr[0].set_title(f"Paciente {key}")
        axarr[0].imshow(img, cmap='gray_r')
        axarr[1].imshow(mask, cmap='gray_r')
        # Add title to the subplots
        axarr[0].set_title(f"Paciente {key}")
        axarr[1].set_title("M√°scara")
        axarr[2].set_title("Overlay")
        
        # Overlay the mask on the image
        axarr[2].imshow(img, cmap='gray_r')
        axarr[2].imshow(mask, cmap='jet', alpha=0.5)

        # Mostrar la figura con subgr√°ficos
        plt.show()

        # Romper el bucle si se alcanza el n√∫mero deseado de muestras
        if samples_counter >= n_samples:
            break
    
    # Romper el bucle si se alcanza el n√∫mero deseado de pacientes
    if patients_counter >= n_patients:
        break
```
